```python
import random
import pdb
ponte={
    'tocha':False,
    'A':False,
    'B':False,
    'C':False,
    'D':False
}
# Definindo os tempos de cada pessoa
times={
    'A':1,
    'B':2,
    'C':5,
    'D':8
}
two_turn = True
t_max=15 # tempo maximo para eles atravesarem a ponte
t_total=0
def mover_uma(tocha, pessoa):
    global t_total, t_max,ponte,times, two_turn
    if ponte[tocha] == ponte[pessoa] and t_total < t_max:
        ponte[tocha] = not ponte[tocha]
        ponte[pessoa] = not ponte[pessoa]
        t_total += times[pessoa]
        print(pessoa, 'retorna com a', tocha, 'levando', t_total, 'minutos')
        two_turn=True
        
def mover_dois(tocha, pessoa1, pessoa2):
    global t_total, t_max, ponte, times, two_turn
    if ponte[tocha] == ponte[pessoa1] == ponte[pessoa2] and t_total < t_max:
        ponte[tocha] = not ponte[tocha]
        ponte[pessoa1] = not ponte[pessoa1]
        ponte[pessoa2] = not ponte[pessoa2]
        t_total += max(times[pessoa1], times[pessoa2])
        print(pessoa1,'e', pessoa2, 'cruzam a ponte com a', tocha, 'levando', t_total, 'minutos')
        two_turn=False
        
def get_list(passou=False):
    global ponte
    lista_passaram=[]
    lista_esperando=[]
    if ponte['A']:
        lista_passaram.append('A')
    else:
        lista_esperando.append('A')
    if ponte['B']:
        lista_passaram.append('B')
    else:
        lista_esperando.append('B')
    if ponte['C']:
        lista_passaram.append('C')
    else:
        lista_esperando.append('C')
    if ponte['D']:
        lista_passaram.append('D')
    else:
        lista_esperando.append('D')
    if passou:
        return lista_passaram
    else:
        return lista_esperando
    
def ponte_final():
    global ponte, t_total,t_max
    if ponte['tocha'] and ponte['A'] and ponte['B'] and ponte['C'] and ponte['D'] and t_total==t_max:
        return True
    else:
        return False
def reset_all():
    global t_total,two_turn,ponte
    t_total=0
    two_turn=True
    ponte['tocha']=ponte['A'] =ponte['B']= ponte['C'] = ponte['D']=False
    print("\n")
    print("Refazendo a solução")
def random_search():
    global t_max,t_total,two_turn
    while not ponte_final():
        reset_all()
        while t_total < t_max:
            if two_turn:
                if t_total == 0:
                    sampled=['A','B']
                else:
                    sampled= random.choices(get_list(), k=2)
                    while sampled[0] == sampled[1]:
                        sampled= random.choices(get_list(), k=2)
                mover_dois('tocha',sampled[0], sampled[1])
            else:
                sampled= random.choices(get_list(True), k=1)
                mover_uma('tocha',sampled[0])
        
random_search()
print(ponte)
```
